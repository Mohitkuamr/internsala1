```python
Ans = 10
import psycopg2
import plotly.graph_objects as go

# Establish a connection to the PostgreSQL database
conn = psycopg2.connect(
    host="your_host",
    database="your_database",
    user="your_username",
    password="your_password"
)

# Create a cursor object to execute database queries
cur = conn.cursor()

# Fetch data from the database
cur.execute("SELECT product_name, sales FROM products")
rows = cur.fetchall()

# Separate the fetched data into separate lists
products = [row[0] for row in rows]
sales = [row[1] for row in rows]

# Close the cursor and database connection
cur.close()
conn.close()

# Create a bar chart using Plotly
fig = go.Figure(data=go.Bar(x=products, y=sales))

# Customize the layout
fig.update_layout(
    title='Sales Performance by Product',
    xaxis_title='Products',
    yaxis_title='Sales',
    barmode='group'
)

# Display the interactive chart
fig.show()
```

```python
Ans = 11
from flask import Flask, jsonify
import psycopg2

app = Flask(__name__)

# Database connection details
DB_HOST = "your_host"
DB_NAME = "your_database"
DB_USER = "your_username"
DB_PASSWORD = "your_password"

# API endpoint to fetch data from the database
@app.route('/data', methods=['GET'])
def get_data():
    try:
        # Establish a connection to the PostgreSQL database
        conn = psycopg2.connect(
            host=DB_HOST,
            database=DB_NAME,
            user=DB_USER,
            password=DB_PASSWORD
        )

        # Create a cursor object to execute database queries
        cur = conn.cursor()

        # Fetch data from the database
        cur.execute("SELECT * FROM your_table")
        rows = cur.fetchall()

        # Close the cursor and database connection
        cur.close()
        conn.close()

        # Convert the database result to a list of dictionaries
        data = []
        for row in rows:
            item = {
                'column1': row[0],
                'column2': row[1],
                # Add more columns as needed
            }
            data.append(item)

        # Return the data as a JSON response
        return jsonify(data)

    except psycopg2.Error as e:
        # Handle any database errors
        return jsonify({'error': str(e)}), 500

if __name__ == '__main__':
    app.run(debug=True)
```
